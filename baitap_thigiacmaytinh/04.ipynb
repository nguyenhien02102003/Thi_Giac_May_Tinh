```python
# Import thư viện
from PIL import Image
from imgtools import *
```

```python
#Đường dẫn ảnh
my_path = 'D:/JS Duy/DNTU/nam4/ki1/thigiacmaytinh/img/01/01.jpg'
```

```python
# Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 75, 175, 150)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán 
# Xác định vị trí cần dán
paste_position = (150, 150)
img.paste(cropped_image, paste_position)
display(img)
```
