```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Đường dẫn ảnh
my_path = 'D:/JS Duy/DNTU/nam4/ki1/thigiacmaytinh/img/01/01.jpg'
#Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyển sang ảnh xám 
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# Tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=50)
plt.title('Biểu đồ histogram của ảnh')
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# Làm nổi bất đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, origin='image')
plt.show()
```

```python
# Duong Dan Anh
my_path = 'D:/JS Duy/DNTU/nam4/ki1/thigiacmaytinh/img/01/02.jpg'

# Doc Anh va Hien Thi Anh Vs Truc Toa Do
im = Image.open(my_path).convert('L')

# Hien thi anh tren truc toa do
plt.imshow(im)

# Tao Hinh Ve Moi
plt.figure()

# Dat Mau Sac La Mau Xam
plt.gray()

# Ve Bieu Do Duong Bien (contour) cua anh
plt.contour(im, origin="image")

# Dam bao ty le giua truc x va truc y
plt.axis('equal')

# Hien Thi Anh
plt.show()
```
